```python
import pandas as pd
import seaborn as sns
```

```python
path_covid_19_clean = r'C:\Users\angel\Desktop\Data science\GitHub\GitHub\btc2-blue-g4\src\Dados\kaggle\covid_19_clean_complete.csv'
path_worldmeter = r'C:\Users\angel\Desktop\Data science\GitHub\GitHub\btc2-blue-g4\src\Dados\kaggle\worldometer_data.csv'
path_country_wise = r'C:\Users\angel\Desktop\Data science\GitHub\GitHub\btc2-blue-g4\src\Dados\kaggle\country_wise_latest.csv'
path_full_grouped = r'C:\Users\angel\Desktop\Data science\GitHub\GitHub\btc2-blue-g4\src\Dados\kaggle\full_grouped.csv'
```

```python
def gerar_df(path_covid_19_clean,path_worldmeter, path_country_wise, path_full_grouped):

    df0 = pd.read_csv(path_covid_19_clean)
    df1 = pd.read_csv(path_worldmeter)
    df2 = pd.read_csv(path_full_grouped )

    df3 = pd.merge(df0, df1[["Continent","Population","Country/Region"]], on=["Country/Region"], how="left")
    
    df_diario = pd.merge(df3, df2[["New cases","New deaths","Date", "Country/Region"]], on=["Date","Country/Region" ], how="left")

    df_diario= df_diario.drop(["Lat","Long", 
                   "Province/State", 
                   "WHO Region"], axis=1)

    df_diario["death / 100.000 Hab"] = (df_diario["Deaths"]/df_diario["Population"])*100000

    df_diario = df_diario[["Country/Region",
               "Continent", 
               "Population", 
               "Date", 
               "Confirmed", 
               "Deaths", 
               "death / 100.000 Hab", 
               "Recovered", "Active", 
               "New cases", 
               "New deaths"]]

  
    df_diario.rename(columns={
    'Province/State': 'province_state',
    'Country/Region': 'country_region',
    },inplace=True)
    df_diario.columns = df_diario.columns.str.lower()
    df_diario["date"] = pd.to_datetime(df_diario["date"])


    df_total = pd.read_csv(path_country_wise)

    df_total = pd.merge(df_total, df1[["Continent","Population","Country/Region"]], on=["Country/Region"], how="left")
    df_total["death / 100.000 Hab"] = (df_total["Deaths"]/df_total["Population"])*100000

    df_total= df_total.drop([ "New cases",
                              "New deaths", 
                              "New recovered", 
                              "WHO Region",
                               "Confirmed last week", 
                               "1 week change", 
                               "1 week % increase"], axis=1)

    df_total = df_total[["Country/Region",
                        "Continent", 
                        "Population", 
                        "Confirmed", 
                        "Deaths", 
                        "death / 100.000 Hab", 
                        "Recovered", 
                        "Active",]]

    df_total.rename(columns={
    'Province/State': 'province_state',
    'Country/Region': 'country_region',
    },inplace=True)
    df_total.columns = df_total.columns.str.lower()

    df_total.reset_index()
    df_diario.reset_index()

           
    return df_total, df_diario
    
```

```python
df_total, df_diario = gerar_df(path_covid_19_clean,path_worldmeter, path_country_wise, path_full_grouped)
```
